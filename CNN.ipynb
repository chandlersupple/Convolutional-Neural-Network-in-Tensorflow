```
import tensorflow as tf
sess = tf.InteractiveSession()
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data", one_hot= True)
```

```
x = tf.placeholder(tf.float32, shape= [None, 784])
y = tf.placeholder(tf.float32, shape= [None, 10])
```

```
def weight_init(shape):
  values = tf.truncated_normal(shape= shape, stddev= 0.1)
  return tf.Variable(values)
def bias_init(shape):
  values = tf.constant(0.1, shape= shape)
  return tf.Variable(values)
def conv2D(x, W):
  return tf.nn.conv2d(x, W, strides= [1, 1, 1, 1], padding= 'SAME')
def max_pool(x):
  return tf.nn.max_pool(x, ksize= [1, 2, 2, 1], strides= [1, 2, 2, 1], padding= 'SAME')
```

```
x_4D = tf.reshape(x, [-1, 28, 28, 1])

conv_W1 = weight_init([5, 5, 1, 32])
conv_b1 = bias_init([32])

conv_op1 = tf.nn.relu(conv2D(x_4D, conv_W1) + conv_b1)
mp_op1 = max_pool(conv_op1)
```

```
conv_W2 = weight_init([5, 5, 32, 64])
conv_b2 = bias_init([64])

conv_op2 = tf.nn.relu(conv2D(mp_op1, conv_W2) + conv_b2)
mp_op2 = max_pool(conv_op2)
```

```
flatten = tf.reshape(mp_op2, [-1, 7  * 7 * 64])
```

```
fc_W1 = weight_init([7 * 7 * 64, 1024])
fc_b1 = bias_init([1024])

fc_op1 = tf.nn.relu(tf.matmul(flatten, fc_W1) + fc_b1)
dropout1 = tf.nn.dropout(fc_op1, keep_prob= 0.5)
```

```
fc_W2 = weight_init([1024, 10])
fc_b2 = bias_init([10])

fc_op2 = tf.matmul(dropout1, fc_W2) + fc_b2
```

```
cross_entropy = tf.nn.softmax_cross_entropy_with_logits_v2(labels= y, logits= fc_op2)
optimize = tf.train.AdadeltaOptimizer(0.001).minimize(cross_entropy)
```

```
correct_predictions = tf.equal(tf.argmax(fc_op2 ,1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_predictions, tf.float32))
```

```
sess.run(tf.global_variables_initializer())

for step in range (100000):
  batch = mnist.train.next_batch(100)
  optimize.run(feed_dict= {x: batch[0], y: batch[1]})
  if step % 1000 == 0:
    step_accuracy = accuracy.eval(feed_dict= {x: batch[0], y: batch[1]})
    print("Step %s, Accuracy %s" %(step, step_accuracy))
```

```
test_accuracy = accuracy.eval(feed_dict= {x: mnist.test.images, y: mnist.test.labels})
print("Test Accuracy: %s" %(test_accuracy))
```
